```python
import matplotlib.pyplot as plt
import numpy as np
from cobaya.model import get_model
from tqdm import tqdm
```

```python
taus = np.linspace(0.035, 0.08, 121)

info = {
    "likelihood": {"elica": None},
    "params": {
        "As": {
            "latex": "A_\\mathrm{s}",
            "value": "lambda tau: 1.884 * 1e-9 * np.exp(2 * tau)",
        },
        "H0": 67.32,
        "mnu": 0.06,
        "ns": 0.9651,
        "ombh2": 0.02237,
        "omch2": 0.1201,
        "tau": 0.0544,
    },
    "theory": {
        "camb": {
            "extra_args": {
                "lens_potential_accuracy": 1,
                "nnu": 3.044,
                "num_massive_neutrinos": 1,
                "theta_H0_range": [20, 100],
            }
        }
    },
}
```

```python
loglikes = []
for i in tqdm(range(len(taus))):
    info["params"]["tau"] = taus[i]
    model = get_model(info)
    loglike = model.loglikes({})[0][0]
    loglikes.append(loglike)
loglikes = np.array(loglikes)
norm_loglikes = loglikes - np.max(loglikes)
```

```python
plt.plot(taus, np.exp(norm_loglikes))

peak = taus[np.argmax(loglikes)]
plt.vlines(peak, 0, 1, colors="red", linestyles="--", label=f"Peak at tau={peak:.4f}")
plt.legend()

plt.xlabel("tau")
plt.ylabel("exp(loglike - max(loglike))")
plt.grid()

plt.xlim(0.04, 0.08)

plt.show()
```
